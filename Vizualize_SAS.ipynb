```
plt.figure(figsize=(14,14))
map = Basemap(llcrnrlon=40, llcrnrlat=0, urcrnrlon=120, urcrnrlat=40,projection='mill')

map.drawrivers(color='lightblue')
map.drawcountries(color='gray')
map.shadedrelief()
map.drawmapboundary()

#Pakistan
p_x = 69.3451
p_y = 30.3753

#Bangladesh
b_x = 90.3563
b_y = 23.6850

#Sri-lanka
s_x = 80.7718
s_y = 7.8731

#ITY
i_x = 82.7718
i_y = 19.8731

#GIH
g_x = 69.3451
g_y = 23.6850

x, y = map(p_x, p_y)
plt.text(x - 230000, y, 'PJL',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x, y, 'ro')

x1, y1 = map(b_x, b_y)
plt.text(x1 + 300000, y1, 'BEB',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x1, y1, 'ro')

x2, y2 = map(s_x, s_y)
plt.text(x2 - 200000, y2 - 240000, 'STU',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x2, y2, 'ro')

x3, y3 = map(i_x, i_y)
plt.text(x3 + 300000, y3 - 100000, 'ITU',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x3, y3, 'ro')

x4, y4 = map(g_x, g_y)
plt.text(x4 - 230000, y4, 'GIH',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x4, y4, 'ro')

plt.plot([x, x3], [y, y3], linewidth=2.0, color = px.colors.sequential.Plasma[3])
plt.plot([x2, x3], [y2, y3], linewidth=2.0, color = px.colors.sequential.Plasma[3])
plt.plot([x3, x1], [y3, y1], linewidth=2.0, color = px.colors.sequential.Plasma[0])
plt.plot([x4, x3], [y4, y3], linewidth=2.0, color = px.colors.sequential.Plasma[3])

plt.text((x + x3)/2 + 200000, (y + y3)/2, '0.78', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x2 + x3)/2 - 240000 , (y2 + y3)/2, '0.78', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x3 + x1)/2 - 200000, (y3 + y1)/2, '0.75', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x4 + x3)/2 + 80000, (y4 + y3)/2, '0.78', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')

plt.xlabel('Точность бинарной классификаций популяций Южной Азии относительно популяции ITU')

plt.legend(fontsize=20, shadow=True, framealpha=1, facecolor='y', edgecolor='r', title='Расшифровка аббревиатур \n популяции:\n \n PJL - Пакистан \n BEB - Бангладеш \n STU - Шри-Ланка \n ITU - Андхра-Прадеш \n GIH - Хьюстон', loc = 'lower right')

plt.savefig('ITU.jpg', dpi = 200)
```
