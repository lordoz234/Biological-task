**CEU**

```
plt.figure(figsize=(14,14))
map = Basemap(llcrnrlon=-30, llcrnrlat=30, urcrnrlon=56, urcrnrlat=65,projection='mill')

map.drawrivers(color='lightblue')
map.drawcountries(color='gray')
map.shadedrelief()
map.drawmapboundary()

#Spain
s_x = -3.7
s_y = 40.4

#France
f_x = 2.2
f_y = 46.2

#Italy
i_x = 12.5674
i_y = 41.8719

#UK
u_x = -3.4360
u_y = 55.3781

#Finalnd
fi_x = 25.7482
fi_y = 61.9241

x, y = map(s_x, s_y)
plt.text(x - 200000, y, 'IBS',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x, y, 'ro')

x1, y1 = map(f_x, f_y)
plt.text(x1 - 350000, y1, 'CEU',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x1, y1, 'ro')

x2, y2 = map(i_x, i_y)
plt.text(x2 + 250000, y2, 'TSI',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x2, y2, 'ro')

x3, y3 = map(u_x, u_y)
plt.text(x3 + 280000, y3, 'GBR',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x3, y3, 'ro')

x4, y4 = map(fi_x, fi_y)
plt.text(x4 + 280000, y4, 'FIN',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x4, y4, 'ro')

plt.plot([x, x1], [y, y1], linewidth=2.0, color = px.colors.sequential.Plasma[2])
plt.plot([x2, x1], [y2, y1], linewidth=2.0, color = px.colors.sequential.Plasma[0])
plt.plot([x3, x1], [y3, y1], linewidth=2.0, color = px.colors.sequential.Plasma[0])
plt.plot([x4, x1], [y4, y1], linewidth=2.0, color = px.colors.sequential.Plasma[3])

plt.text((x + x1)/2 + 300000, (y + y1)/2, '0.76', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x2 + x1)/2 + 150000, (y2 + y1)/2, '0.74', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x3 + x1)/2 + 200000, (y3 + y1)/2, '0.74', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x4 + x1)/2 - 200000, (y4 + y1)/2, '0.78', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')

plt.xlabel('Точность бинарной классификаций популяций Европы относительно популяции CEU')

plt.legend(fontsize=20, shadow=True, framealpha=1, facecolor='y', edgecolor='r', title='Расшифровка аббревиатур \n популяции:\n \n GBR - Британия \n FIN - Финляндия \n CEU - Юта \n IBS - Испания \n TSI - Италия', loc = 'lower right')

plt.savefig('CEU.jpg', dpi = 200)
```
