**CHS**

```
plt.figure(figsize=(14,14))
map = Basemap(llcrnrlon=80, llcrnrlat=10, urcrnrlon=150, urcrnrlat=45,projection='mill')

map.drawrivers(color='lightblue')
map.drawcountries(color='gray')
map.shadedrelief()
map.drawmapboundary()

#CHS
s_x = 113.2644
s_y = 23.1291

#CDX
x_x = 100.7971
x_y = 22.0088

#KHV
k_x = 108.2772
k_y = 14.0583

#JPT
j_x = 138.2529
j_y = 36.2048

#CHB
b_x = 116.4074
b_y = 39.9042

x, y = map(s_x, s_y)
plt.text(x + 230000, y - 100000, 'CHS',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x, y, 'ro')

x1, y1 = map(x_x, x_y)
plt.text(x1 - 300000, y1, 'CDX',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x1, y1, 'ro')

x2, y2 = map(k_x, k_y)
plt.text(x2 - 200000, y2 - 220000, 'KHV',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x2, y2, 'ro')

x3, y3 = map(j_x, j_y)
plt.text(x3 + 260000, y3 - 100000, 'JPT',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x3, y3, 'ro')

x4, y4 = map(b_x, b_y)
plt.text(x4 - 230000, y4, 'CHB',fontsize=13,fontweight='bold',va='bottom', ha='center', color='black')
plt.plot(x4, y4, 'ro')

plt.plot([x1, x], [y1, y], linewidth=2.0, color = px.colors.sequential.Plasma[0])
plt.plot([x2, x], [y2, y], linewidth=2.0, color = px.colors.sequential.Plasma[1])
plt.plot([x3, x], [y3, y], linewidth=2.0, color = px.colors.sequential.Plasma[2])
plt.plot([x, x4], [y, y4], linewidth=2.0, color = px.colors.sequential.Plasma[2])

plt.text((x1 + x)/2 + 220000, (y1 + y)/2 + 50000, '0.72', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x2 + x)/2 + 200000, (y2 + y)/2, '0.73', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x3 + x)/2 - 150000, (y3 + y)/2, '0.74', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')
plt.text((x + x4)/2 - 170000, (y + y4)/2 + 60000, '0.74', fontsize = 13, fontweight='ultralight',va='bottom', ha='center', color='black')

plt.xlabel('Точность бинарной классификаций популяций Восточной Азии относительно популяции CHS')

plt.legend(fontsize=20, shadow=True, framealpha=1, facecolor='y', edgecolor='r', title='Расшифровка аббревиатур \n популяции:\n \n CHS - Гуанчжоу \n CDX - Юньнань \n KHV - Вьетнам \n CHB - Пекин \n JPT - Япония', loc = 'lower right')

plt.savefig('CHS.jpg', dpi = 200)
```
